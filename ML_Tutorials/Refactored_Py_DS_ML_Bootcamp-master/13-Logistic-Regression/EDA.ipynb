```python
def fill_age(col):
    age_col=col[0]
    class_col=col[1]
    if pd.isnull(age_col):
        if class_col==1:
            return 50
        elif class_col==2:
            return 35
        else:
            return 20
    else:
        return age_col
```

```python
test['Age']=test[['Age','Pclass']].apply(fill_age,axis=1)
```
